```python
# Dependencies
import os
import csv
import codecs
import pandas as pd
```

```python
# Specify the file path
thisFile_path = os.path.abspath("")
csvpath_schRawData = os.path.join(thisFile_path, "School Rankings_RAW_2018.csv")

print(csvpath_schRawData)
```

```python
# Read the CSV file with csv.reader
with codecs.open(csvpath_schRawData, 'r', encoding='UTF-8', errors='ignore') as csvfile:
#with open(csvpath_schRawData, 'r', encoding='UTF-8') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=" ")
    header = next(csvreader)
    data = [row for row in csvreader]  # Collect the data rows

# Convert csvreader object and header into a DataFrame
df = pd.DataFrame.from_records(data, columns=header)

# Write the DataFrame to CSV
csvpath_sch_interm = os.path.join(thisFile_path, "School Rankings_interm_2018.csv")
df.to_csv(csvpath_sch_interm, index=False)
```
