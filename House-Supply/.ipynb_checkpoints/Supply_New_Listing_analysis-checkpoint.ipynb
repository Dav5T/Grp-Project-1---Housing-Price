```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the House data
data ="Resources/2018-2021 Housing Price.csv"
data_df = pd.read_csv(data, encoding = "utf-8")
data_df
```

```python
#Create House listing table
listing_df = data_df[['Municipality', 'Total New Listings 2021', 'Total New Listings 2020', 'New Listing 2019', 'New Listings']]

#Rename column
rename_listing_df = listing_df.rename(columns = {'Total New Listings 2021': 2021,
                                                 'Total New Listings 2020': 2020,
                                                 'New Listing 2019': 2019,
                                                'New Listings': 2018})

rename_listing_df
```

```python
#Group by Municipality, and sum new listing total
group = rename_listing_df.groupby('Municipality')

new_table = group.sum()

new_table
```

```python
#draw line chart
new_graph=new_table.plot()

#Set Title and Label
plt.title('Total New Listings')
plt.xlabel('Municipality')
plt.ylabel('Number of New Listings')

new_graph.figure.savefig('Graphs/total_new_listing.png')

plt.show()
```
